```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

LINE CHART

```python
#line chart

data = {
    "age": [64, 21, 30, 35, 39, 68, 50, 23, 34, 36, 39],
    "SysBP": [141, 109, 112, 106, 140, 115, 125, 111, 144, 113, 146]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create a line plot for Age vs Systolic Blood Pressure
plt.plot(df["age"], df["SysBP"], marker='o', linestyle='-', color='b')

# Add title and labels
plt.title("Age vs Systolic Blood Pressure (SysBP)")
plt.xlabel("Age (Years)")
plt.ylabel("Systolic Blood Pressure (mmHg)")

# Show the plot
plt.show()
```

```python
#line chart


data = {
    "age": [64, 21, 30, 35, 39, 68, 50, 23, 34, 36, 39],
    "DiaBP": [96, 100, 73, 90, 90, 93, 117, 85, 80, 109, 104]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create a line plot for Age vs Diastolic Blood Pressure
plt.plot(df["age"], df["DiaBP"], marker='o', linestyle='-', color='g')

# Add title and labels
plt.title("Age vs Diastolic Blood Pressure (DiaBP)")
plt.xlabel("Age (Years)")
plt.ylabel("Diastolic Blood Pressure (mmHg)")

# Show the plot
plt.show()
```

BOX PLOT

```python
#age distribution by sex and age

data = {
    'age': [39, 25, 33, 58, 54, 68, 44, 56, 60],
    'sex': [1, 1, 0, 1, 0, 1, 0, 1, 1],
    'indication': [0, 1, 0, 1, 0, 0, 1, 1, 1]
}
df = pd.DataFrame(data)

df['sex'] = df['sex'].map({1: 'Male', 0: 'Female'})
df['indication'] = df['indication'].map({1: 'Yes', 0: 'No'})

sns.boxplot(x='sex', y='age', hue='indication', data=df)
plt.title('Age Distribution by Sex and Indication')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
data = {  'age': [39, 25, 33, 58, 54, 68, 44, 56, 60],}
df = pd.DataFrame(data)

sns.displot(df['age'], kind="hist", kde=True, color='skyblue')
plt.title('Distribution of Age with KDE')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#bivarient

data = {'age': [39, 25, 33, 58, 54, 68, 44, 56, 60],
    'indication': [0, 1, 0, 1, 0, 0, 1, 1, 1]

}
df = pd.DataFrame(data)

sns.kdeplot(df[df['indication'] == 0]['age'], label='No Indication', color='red')
sns.kdeplot(df[df['indication'] == 1]['age'], label='Indication', color='blue')

plt.title('Age Distribution by Indication')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

HEAT MAP


```python
#heatmap

# Dataset
data = {
    "age": [64, 21, 30, 35, 39, 68, 50, 23, 34, 36, 39],
    "SysBP": [141, 109, 112, 106, 140, 115, 125, 111, 144, 113, 146],
    "DiaBP": [96, 100, 73, 90, 90, 93, 117, 85, 80, 109, 104],
    "HR": [128, 106, 126, 130, 112, 130, 89, 131, 90, 187, 85],
    "weightKg": [69, 48, 69, 45, 92, 53, 56, 55, 67, 46, 54],
    "heightCm": [147, 150, 183, 149, 166, 172, 172, 169, 178, 114, 160]
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)

# Add title
plt.title("Heatmap of Correlations")
plt.show()
```
